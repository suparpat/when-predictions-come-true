## Calculate number of tweets per day

```python
import matplotlib.pyplot as plt
import matplotlib
from datetime import datetime, date, time
from pandas import DataFrame, read_csv
# Enable inline plotting
%matplotlib inline
```

### Read in data

```python
data = read_csv("cleaned_sorted_coke.csv")
data_timestamp = data['ts']
```

### Check that dates are ordered

```python
DateListSentiment = []
for x in data.ts:
    #Format: Mon Jan 13 20:01:56 +0000 2014
    DateListSentiment.append(datetime.strptime(x, "%a %b %d %H:%M:%S +0000 %Y"))
    
plt.plot(DateListSentiment)
plt.ylabel('Dates')  
plt.xticks(rotation=45)
```

### Count number of tweets per day

```python
day = []
tweets = []
count = len(DateListSentiment)

temp = 0

for t in range(1,count):
    #If in the same hour, add polarity to temp
    if DateListSentiment[t-1].day==DateListSentiment[t].day and DateListSentiment[t-1].month==DateListSentiment[t].month and DateListSentiment[t-1].year==DateListSentiment[t].year:
        temp = temp + 1
        
    #If temp is not empty, add average polarity to hourly_sentiments    
    elif temp!=0:
        temp = temp + 1
        day.append('{} {} {}'.format(DateListSentiment[t-1].day,DateListSentiment[t-1].month,DateListSentiment[t-1].year))
        tweets.append(temp)
        temp = 0

combined = zip(day,tweets)

dfday = DataFrame(data = combined, columns = ['day','tweets'])
dfday.to_csv('coke_dailytweets.csv',index=False)
```

### Plot to see number of tweets per day

```python
DateListSentiment3 = []
for d in dfday.day:
    #Format: 
    DateListSentiment3.append(datetime.strptime(d, "%d %m %Y"))
    
dateday = matplotlib.dates.date2num(DateListSentiment3)
plt.plot_date(dateday, dfday.tweets,'bo-')
plt.xlabel('Dates') 
plt.ylabel('Tweets')
plt.xticks(rotation=45)
```
