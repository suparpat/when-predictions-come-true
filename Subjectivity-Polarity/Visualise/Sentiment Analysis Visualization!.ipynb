```python
import matplotlib.pyplot as plt
import matplotlib
from datetime import datetime, date, time
from pandas import DataFrame, read_csv
# Enable inline plotting
%matplotlib inline
```

### Read stock data

```python
stock_df = read_csv("boeing_stockprices.csv")
stock_date = stock_df['Date']
stock_closingprices = stock_df['Close']
```

### Read sentiment data

```python
sentiment_df = read_csv("sorted_polarity.csv")
sentiment_timestamp = sentiment_df['ts']
sentiments_polarity = sentiment_df['polarity']
```

### Prepare data

```python
stock_toplot= zip(stock_date,stock_closingprices)
stocks = DataFrame(data= stock_toplot, columns = ['Date','Close'])

sentiment_toplot= zip(sentiment_timestamp,sentiments_polarity)
sentiments = DataFrame(data = sentiment_toplot, columns = ['ts','polarity'])
```

### Plot: Stock close prices

```python
DateList = []
for x in stocks.Date:
    DateList.append(datetime.strptime(x, "%Y-%m-%d"))
    
dates = matplotlib.dates.date2num(DateList)
plt.plot_date(dates, stocks.Close)
```

### Plot: Sentiments

```python
DateListSentiment = []
for x in sentiments.ts:
    #Format: Mon Jan 13 20:01:56 +0000 2014
    DateListSentiment.append(datetime.strptime(x, "%a %b %d %H:%M:%S +0000 %Y"))
    
timestamps = matplotlib.dates.date2num(DateListSentiment)
plt.plot_date(timestamps, sentiments.polarity)
DateListSentiment[0]
```

### Check if the dates are sorted

```python
plt.plot(DateListSentiment)
```

### Calculate hourly average for the sentiments.

```python
hour = []
hourly_sentiments = []
count = len(DateListSentiment)

temp = []

for t in range(1,count):
    #If in the same hour, add polarity to temp
    if DateListSentiment[t-1].hour==DateListSentiment[t].hour and DateListSentiment[t-1].day==DateListSentiment[t].day and DateListSentiment[t-1].month==DateListSentiment[t].month and DateListSentiment[t-1].year==DateListSentiment[t].year:
        temp.append(sentiments.polarity[t-1])
        
    #If temp is not empty, add average polarity to hourly_sentiments    
    elif len(temp)!=0:
        temp.append(sentiments.polarity[t-1])
        count_temp = len(temp)
        hour.append('{} {} {} {}'.format(DateListSentiment[t-1].day,DateListSentiment[t-1].month,DateListSentiment[t-1].year,DateListSentiment[t-1].hour))
        hourly_sentiments.append(sum(temp)/count_temp)
        temp = []

combined = zip(hour,hourly_sentiments)

dfx = DataFrame(data = combined, columns = ['hour','hourly_sentiments'])
dfx.to_csv('combined.csv',index=False)
```

### Plot: Hourly sentiments

```python
DateListSentiment2 = []
for h in dfx.hour:
    #Format: 
    DateListSentiment2.append(datetime.strptime(h, "%d %m %Y %H"))
    
datex = matplotlib.dates.date2num(DateListSentiment2)
plt.plot_date(datex, dfx.hourly_sentiments)
plt.show()
```

### Calculate daily average for the sentiments

```python
day = []
daily_sentiments = []
count = len(DateListSentiment)

temp = []

for t in range(1,count):
    #If in the same hour, add polarity to temp
    if DateListSentiment[t-1].day==DateListSentiment[t].day and DateListSentiment[t-1].month==DateListSentiment[t].month and DateListSentiment[t-1].year==DateListSentiment[t].year:
        temp.append(sentiments.polarity[t-1])
        
    #If temp is not empty, add average polarity to hourly_sentiments    
    elif len(temp)!=0:
        temp.append(sentiments.polarity[t-1])
        count_temp = len(temp)
        day.append('{} {} {}'.format(DateListSentiment[t-1].day,DateListSentiment[t-1].month,DateListSentiment[t-1].year))
        daily_sentiments.append(sum(temp)/count_temp)
        temp = []

combined = zip(day,daily_sentiments)

dfday = DataFrame(data = combined, columns = ['day','daily_sentiments'])
dfday.to_csv('combined_daily.csv',index=False)
```

### Plot: Daily sentiments

```python
DateListSentiment3 = []
for h in dfday.day:
    #Format: 
    DateListSentiment3.append(datetime.strptime(h, "%d %m %Y"))
    
dateday = matplotlib.dates.date2num(DateListSentiment3)
plt.plot_date(dateday, dfday.daily_sentiments)

```
