```python
import matplotlib.pyplot as plt
import matplotlib
from datetime import datetime, date, time
from pandas import DataFrame, read_csv
# Enable inline plotting
%matplotlib inline
```

### Read stock data

```python
stock_df = read_csv("boeing_stockprices.csv")
stock_date = stock_df['Date']
stock_closingprices = stock_df['Close']
```

### Read sentiment data

```python
sentiment_df = read_csv("polarity.csv")
sentiment_timestamp = sentiment_df['ts']
sentiments_polarity = sentiment_df['polarity']
```

### Prepare data

```python
stock_toplot= zip(stock_date,stock_closingprices)
stocks = DataFrame(data= stock_toplot, columns = ['Date','Close'])

sentiment_toplot= zip(sentiment_timestamp,sentiments_polarity)
sentiments = DataFrame(data = sentiment_toplot, columns = ['ts','polarity'])
```

### Plot!

```python
DateList = []
for x in stocks.Date:
    DateList.append(datetime.strptime(x, "%Y-%m-%d"))
    
dates = matplotlib.dates.date2num(DateList)
plt.plot_date(dates, stocks.Close)
```

```python
DateListSentiment = []
for x in sentiments.ts:
    #Format: Mon Jan 13 20:01:56 +0000 2014
    DateListSentiment.append(datetime.strptime(x, "%a %b %d %H:%M:%S +0000 %Y"))
    
timestamps = matplotlib.dates.date2num(DateListSentiment)
plt.plot_date(timestamps, sentiments.polarity)
```

### Hmmm.. calculate hourly average for the sentiments.

```python
hour = []
hourly_sentiments = []
count = len(DateListSentiment)

temp = []

for t in range(1,count):
    if DateListSentiment[t-1].hour==DateListSentiment[t].hour and DateListSentiment[t-1].day==DateListSentiment[t].day and DateListSentiment[t-1].month==DateListSentiment[t].month and DateListSentiment[t-1].year==DateListSentiment[t].year:
        temp.append(sentiments.polarity[t-1])
        
    elif len(temp)!=0:
        count_temp = len(temp)
        hour.append('{} {} {} {}'.format(DateListSentiment[t-1].hour,DateListSentiment[t-1].day,DateListSentiment[t-1].month,DateListSentiment[t-1].year))
        hourly_sentiments.append(sum(temp)/count_temp)
        temp = []

combined = zip(hour,hourly_sentiments)

dfx = DataFrame(data = combined, columns = ['hour','hourly_sentiments'])
dfx.to_csv('combined.csv',index=False)
```

### Plot again!

```python
DateListSentiment2 = []
for h in dfx.hour:
    
    #Format: 
    DateListSentiment2.append(datetime.strptime(h, "%H %d %m %Y"))
    
datex = matplotlib.dates.date2num(DateListSentiment2)
plt.plot_date(datex, dfx.hourly_sentiments)
plt.show()
```
